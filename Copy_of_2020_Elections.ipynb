```python
df = pd.read_csv('2020_elections.csv')
```

```python
def convert(array):
    lst = []
    for num in array:
        string = ''
        if len(num)<=7:
            string=num[-3:]+string
            string=num[-7:-4]+string
        else:
            string=num[-3:]+string
            string=num[-7:-4]+string
            string=num[:-8]+string
        lst.append(int(string))
    return lst
```

```python
df['Voting-Eligible Population (VEP)'] = convert(list(df['Voting-Eligible Population (VEP)']))
```

```python
df['Total Ineligible Felon'] = convert(list(df['Total Ineligible Felon']))
```

```python
def average(lst):
  sum = 0
  for i in lst:
    sum+=i
  return sum/len(lst)
```

```python
def convertPercent(lst):
  for i in range(0,len(lst)):
    lst[i] = float(lst[i][:-1])
  return lst
```
